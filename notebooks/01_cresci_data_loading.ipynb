```python
import pandas as pd
from pathlib import Path
from zipfile import ZipFile

data_dir = Path("../data/raw/cresci_2017")
zip_files = sorted(data_dir.glob("*.zip"))

schema_info = []
base_columns = None

for zip_path in zip_files:
    with ZipFile(zip_path, "r") as z:
        user_files = [
            name for name in z.namelist()
            if not name.startswith("__MACOSX") and name.endswith("users.csv")
        ]
        
        if not user_files:
            continue
        
        user_file = user_files[0]
        df_sample = pd.read_csv(z.open(user_file), nrows=0)
        cols = df_sample.columns.tolist()
        
        if base_columns is None:
            base_columns = cols
        
        same_as_base = cols == base_columns
        
        schema_info.append({
            "zip_file": zip_path.name,
            "user_file": user_file,
            "column_count": len(cols),
            "same_as_first_file": same_as_base
        })

schema_df = pd.DataFrame(schema_info)
schema_df
```

```python
import pandas as pd
from pathlib import Path
from zipfile import ZipFile

data_dir = Path("../data/raw/cresci_2017")
zip_files = sorted(data_dir.glob("*.zip"))

columns_by_file = {}

for zip_path in zip_files:
    with ZipFile(zip_path, "r") as z:
        user_files = [
            name for name in z.namelist()
            if not name.startswith("__MACOSX") and name.endswith("users.csv")
        ]
        
        if not user_files:
            continue
        
        user_file = user_files[0]
        df_sample = pd.read_csv(z.open(user_file), nrows=0)
        columns_by_file[zip_path.name] = set(df_sample.columns)

all_columns = sorted(set.union(*columns_by_file.values()))
common_columns = sorted(set.intersection(*columns_by_file.values()))

print("Toplam farklı sütun sayısı:", len(all_columns))
print("Tüm dosyalarda ortak sütun sayısı:", len(common_columns))

print("\nTüm dosyalarda ortak sütunlar:")
print(common_columns)

print("\nDosya bazında eksik sütun sayıları:")
for file_name, cols in columns_by_file.items():
    missing_cols = sorted(set(all_columns) - cols)
    print(f"{file_name}: {len(missing_cols)} eksik sütun")
    if missing_cols:
        print("  Eksik olanlar:", missing_cols)
```

```python
import pandas as pd
from pathlib import Path
from zipfile import ZipFile

# 1) Ortak sütunları sabitliyoruz
common_columns = [
    'contributors_enabled', 'created_at', 'default_profile',
    'default_profile_image', 'description', 'favourites_count',
    'follow_request_sent', 'followers_count', 'following',
    'friends_count', 'geo_enabled', 'id', 'is_translator', 'lang',
    'listed_count', 'location', 'name', 'notifications',
    'profile_background_color', 'profile_background_image_url',
    'profile_background_image_url_https', 'profile_background_tile',
    'profile_banner_url', 'profile_image_url',
    'profile_image_url_https', 'profile_link_color',
    'profile_sidebar_border_color', 'profile_sidebar_fill_color',
    'profile_text_color', 'profile_use_background_image',
    'protected', 'screen_name', 'statuses_count', 'time_zone',
    'updated', 'url', 'utc_offset', 'verified'
]

# 2) Hangi dosya bot, hangisi human belirliyoruz
label_map = {
    "genuine_accounts.csv.zip": 0,
    "fake_followers.csv.zip": 1,
    "social_spambots_1.csv.zip": 1,
    "social_spambots_2.csv.zip": 1,
    "social_spambots_3.csv.zip": 1,
    "traditional_spambots_1.csv.zip": 1,
    "traditional_spambots_2.csv.zip": 1,
    "traditional_spambots_3.csv.zip": 1,
    "traditional_spambots_4.csv.zip": 1,
}

# 3) Daha açıklayıcı kaynak etiketi de ekliyoruz
source_map = {
    "genuine_accounts.csv.zip": "genuine_accounts",
    "fake_followers.csv.zip": "fake_followers",
    "social_spambots_1.csv.zip": "social_spambots_1",
    "social_spambots_2.csv.zip": "social_spambots_2",
    "social_spambots_3.csv.zip": "social_spambots_3",
    "traditional_spambots_1.csv.zip": "traditional_spambots_1",
    "traditional_spambots_2.csv.zip": "traditional_spambots_2",
    "traditional_spambots_3.csv.zip": "traditional_spambots_3",
    "traditional_spambots_4.csv.zip": "traditional_spambots_4",
}

data_dir = Path("../data/raw/cresci_2017")
zip_files = sorted(data_dir.glob("*.zip"))

dfs = []

for zip_path in zip_files:
    if zip_path.name == "crowdflower_results.csv.zip":
        continue  # şimdilik kullanmıyoruz

    with ZipFile(zip_path, "r") as z:
        user_files = [
            name for name in z.namelist()
            if not name.startswith("__MACOSX") and name.endswith("users.csv")
        ]

        if not user_files:
            continue

        user_file = user_files[0]

        df_temp = pd.read_csv(
            z.open(user_file),
            usecols=common_columns,
            low_memory=False
        )

        df_temp["source_dataset"] = source_map[zip_path.name]
        df_temp["bot_label"] = label_map[zip_path.name]

        dfs.append(df_temp)

combined_df = pd.concat(dfs, ignore_index=True)

print("Birleşik veri boyutu:", combined_df.shape)
print("\nKaynak veri seti dağılımı:")
print(combined_df["source_dataset"].value_counts())

print("\nBot etiketi dağılımı:")
print(combined_df["bot_label"].value_counts())

combined_df.head()
```

```python
from pathlib import Path

processed_dir = Path("../data/processed")
processed_dir.mkdir(parents=True, exist_ok=True)

output_path = processed_dir / "cresci_2017_users_combined.csv"

combined_df.to_csv(output_path, index=False)

print("Dosya kaydedildi:", output_path)
print("Kaydedilen veri boyutu:", combined_df.shape)
```
